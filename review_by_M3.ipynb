## Explanation of my M3 code

M3 parses the collated answers that have been derived from the respondent quizzes. First, M3 separates the collated file into sections for individual respondents; then, it applies the M1 parser to turn each respondent section into a series of 100 numerical answers.

The function `generate_means_sequence()` calculates the mean of the answer per question. The value of an unanswered responses is coded as 0; hence, when computing the mean, such questions do not take part since they cannot be considered valid answer options.

The function `visualize_data()` generates two types of plots depending on the value of n. If `n=1`, it produces a scatter plot of the means sequence. If `n=2`, it generates a line plot of all the respondents' sequences.

## Limitations

The limitation is that the mean plot gives information on the behavior of the participants but not the correctness of the solution sequence. Participants could have made errors, guessed, or left the answers blank. Hence, the plots are just exploratory and not definitive of the pattern. The individual plot could get cluttered if many participants are shown at once.
